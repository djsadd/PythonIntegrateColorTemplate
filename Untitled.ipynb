```python
def format_fio(fio: str) -> str:
    def capitalize_part(part):
        return '-'.join(p.capitalize() for p in part.split('-'))
    
    return ' '.join(capitalize_part(word.strip()) for word in fio.strip().split())
```

```python
def format_op_name(op_code):
    if '-' not in op_code:
        return op_code
    
    code, name = op_code.split('-', 1)
    name_lower = name.lower()       # Все строчные
    formatted_name = name_lower.capitalize()  # Первая буква заглавная, остальные строчные
    
    return f"{code}-{formatted_name}"
```

```python
from datetime import datetime

def extract_date_time1(date_str):
    formats = ["%d.%m.%Y %H:%M:%S", "%d.%m.%Y", "%Y-%m-%d %H:%M:%S", "%Y-%m-%d"]  # добавил несколько форматов
    
    dt = None
    for fmt in formats:
        try:
            dt = datetime.strptime(date_str, fmt)
            break
        except ValueError:
            continue
    
    if dt is None:
        raise ValueError("Неверный формат даты")
    
    # Возвращаем строку в формате DD.MM.YYYY
    return dt.strftime("%d.%m.%Y")
```

```python
from datetime import datetime

def extract_date_time(date_str):
    date_str = extract_date_time1(date_str)
    print(date_str)
    formats = ["%d.%m.%Y %H:%M:%S", "%d.%m.%Y"]
    dt = None
    for fmt in formats:
        try:
            dt = datetime.strptime(date_str, fmt)
            break
        except ValueError:
            continue
    
    if dt is None:
        raise ValueError("Неверный формат даты")
    
    day = dt.day
    month = dt.month
    year = dt.year
    
    if "%H" in fmt:
        time = dt.time()  # время (часы, минуты, секунды)
    else:
        time = None
    
    return (day, month, year, time)
```

```python
from PIL import Image
import numpy as np
import cv2

def remove_white_background(input_path: str, output_path: str, threshold: int = 200):
    pil_img = Image.open(input_path).convert("RGBA")
    img = np.array(pil_img)
    
    bgr = cv2.cvtColor(img, cv2.COLOR_RGBA2BGR)
    gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
    
    _, mask = cv2.threshold(gray, threshold, 255, cv2.THRESH_BINARY)

    mask_inv = cv2.bitwise_not(mask)

    img[:, :, 3] = mask_inv

    result_img = Image.fromarray(img)
    result_img.save(output_path)

remove_white_background(
    r"C:\Users\admin\Дипломки\qr-коды\00016635125.png",
    r"C:\Users\admin\Дипломки\без фона\00016635125.png"
)
```
